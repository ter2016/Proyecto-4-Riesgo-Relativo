```python
#Prueba 3
# Seleccionar solo las columnas a analizar
interested_columns = ['age', 'dependents_c', 'last_month_salary_c']
# Agregar la columna 'last_month_salary' al DataFrame existente
#df_subset['last_month_salary_c'] = df['last_month_salary_c']  # Crear un nuevo DataFrame con las columnas seleccionadas
df_subset = df[interested_columns]


# Crear un diccionario para almacenar la cantidad de outliers y los valores ms extremos por variable
outliers_info_dict = {}
cleaned_dfs_dict = {}  #Para almacenar los df limpios por variable

# Iterar sobre las columnas seleccionadas y crear subgrficos para boxplot y histograma
#Ciclo for para dar n vueltas y repetir las acciones dentro del for
for i, column in enumerate(interested_columns, 1):
  # Calcular los percentiles para identificar outliers
  Q1 = df_subset[column].quantile(0.25)
  Q2 = df_subset[column].quantile(0.50)
  Q3 = df_subset[column].quantile(0.75)
  IQR = Q3 - Q1

  # Definir lmites para identificar outliers
  lower_limit = Q1 - 1.5 * IQR
  upper_limit = Q3 + 1.5 * IQR

  # Filtrar datos para identificar outliers
  outliers = df_subset[(df_subset[column] < lower_limit) | (df_subset[column] > upper_limit)]
  #print(outliers)

  # Almacenar informacin sobre outliers y valores ms extremos
  outliers_info_dict[column] = {
      'q1_value': Q1,
      'q2_value': Q2,
      'q3_value': Q3,
      'iqr_value': IQR,
      'lower_limit': lower_limit,
      'upper_limit': upper_limit,
      'outliers_count': len(outliers),
      'min_value': df_subset[column].min(),
      'max_value': df_subset[column].max(),
      'first_outlier_value': outliers[column].min() if not outliers.empty else None,
      'mean_value': df_subset[column].mean(),
      'median_value': df_subset[column].median()
  }

  for variable, valor in outliers_info_dict[column].items():
    print(f'{variable}: {valor}')

  #Plot
  plt.figure(figsize=(12, 4))  # Tamano de la figura, largo por ancho
  sbn.boxplot(x=df_subset[column], data=df_subset)
  plt.title(f'Boxplot variable {column}')
  plt.xlabel(column)
  plt.ylabel('Frecuencia')
  plt.show()

  #Histograma
  plt.figure(figsize=(12, 4))
  plt.hist(df_subset[column], bins=30, color='skyblue', edgecolor='black')
  plt.title(f'Histograma variable {column}')
  plt.xlabel(column)
  plt.ylabel('Frecuencia')
  plt.show()


# # Supongamos que outliers_info_dict es tu diccionario
# for columna, info in outliers_info_dict.items():
#   print(f'Informacin de la columna {columna}:')
#   for variable, valor in info.items():
#       print(f'{variable}: {valor}')
#   print('\n')  # Agrega una lnea en blanco entre las columnas para mayor claridad

df_subset.describe()
```
